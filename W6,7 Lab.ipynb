# Exercise 1: Polynomial Regression [20 Marks]

## Run the following cell to read the California Housing dataset.

```python
import pandas as pd
from sklearn.datasets import fetch_california_housing

# Load the California housing dataset
california = fetch_california_housing()
df = pd.DataFrame(california.data, columns=california.feature_names)
df['PRICE'] = california.target
```

```python
df
```

```python
df.info()
```

## Select all features from the dataset except the price to be your input data. (Convert to array) [1 Mark]
## Select the price to be your target label. (Convert to array and reshape it to be column vector) [1 Mark]

```python
X_df = df.loc[:, df.columns != 'PRICE']

X = X_df.to_numpy()
y = df[['PRICE']].to_numpy()
```

```python
X
```

```python
y
```

## Create second order polynomial features. [4 Marks]

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
X_poly = PolynomialFeatures(degree=2, include_bias=False).fit_transform(X)
```

```python
X_poly.shape
```

```python
X_poly
```

## Use StandardScaler to stabdardize the data [4 Marks]

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_poly = StandardScaler().fit_transform(X_poly)
```

```python
X_poly
```

## Create a LinearRegression model to predict price from the polynomial features [5 Marks]

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_poly, y)
```

```python
predicted_price = lr.predict(X_poly)
```

```python
predicted_price
```

## Evaluate the model using r2_score [1 Mark]

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y,predicted_price)
```

# Exercise 2: (Using Train/Test Split - Overfitting) [20 Marks]

## Create polynomial features with degree 2 [3 Marks]

```python
X_poly = PolynomialFeatures(degree=2, include_bias=False).fit_transform(X)
```

## Train/test split using test_size=0.33 and random_state=42 [4 Marks]

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.33, random_state=42)
```

## Scale the training data using StandardScaler [3 Marks]

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

## Use the scaled data to train a linear regression model. [3 Marks]

```python
final_lr = LinearRegression()
final_lr.fit(X_train_scaled, y_train)
```

## Scale the test data using the trained StandardScaler. [2 Marks]

```python
X_test_scaled = scaler.transform(X_test)
```

## Use the trained model to predict the target label for both scaled training and test data. [2 Marks]

```python
y_train_pred = final_lr.predict(X_train_scaled)
y_test_pred = final_lr.predict(X_test_scaled)
```

## Evaluate the predictions using r2_score. [1 Mark]

```python
r2_train_score = r2_score(y_train, y_train_pred)
r2_test_score = r2_score(y_test, y_test_pred)
print('r2_score Train:', r2_train_score)
print('r2_score Test:', r2_test_score)
```

## Repeat using third order polynomial and notice the test scoe [2 Marks]
