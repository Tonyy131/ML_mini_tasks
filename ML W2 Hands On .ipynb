## Simple Linear Regression [10 Marks]

### Use midparentHeight features to create a linear regresion model to predict childHeight

#### 1. Extract X & y from the data (X is midparentHeight and y is the childHeight) [2 Marks]

```python
np.set_printoptions(precision=2, suppress=True)
```

```python
X = df[['midparentHeight']].values
print(repr(X))
```

```python
y = df['childHeight'].values
print(repr(y))
```

#### 2. Plot X vs. y [2 Marks]

```python
plt.scatter(X, y, label='Actual Data', color='blue')
plt.xlabel('Midparent Height')
plt.ylabel('Child Height')
plt.title('Linear Regression: Child Height vs Midparent Height')
plt.legend()
plt.show()
```

#### 3. Create the model [2 Marks]

```python
from sklearn.linear_model import LinearRegression
```

```python
reg_ice = LinearRegression()
reg_ice.fit(X, y)
```

#### 4. Use the model to make prediction [2 Marks]

```python
print("Intercept:", reg_ice.intercept_)
print("Coefficient:", reg_ice.coef_[0])
```

```python
y_pred = reg_ice.predict(X)
print(repr(y_pred))
```

#### 5. Calculate RSS [2 Marks]

```python
RSS = np.sum((y - y_pred) ** 2)
print(f"RSS: {RSS:.2f}")
```

#### 6. Plot X vs. y and the Model Prediction

```python
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('Midparent Height')
plt.ylabel('Child Height')
plt.title('Linear Regression: Child Height vs Midparent Height')
plt.legend()
plt.show()
```

## Multiple Linear Regression [10 Marks]

### Use father,	mother, and	midparentHeight features to create a linear regresion model to predict childHeight

#### Create the model [8 Marks]

```python
np.set_printoptions(precision=2, suppress=True)
```

```python
X_multi = df[['mother', 'father', 'midparentHeight']].values
print(repr(X_multi))
```

```python
y = df['childHeight'].values
print(repr(y))
```

```python
model_multi = LinearRegression()
```

```python
model_multi.fit(X_multi, y)
```

#### Calculate RSS [2 Marks]

```python
y_pred_multi = model_multi.predict(X_multi)
RSS_multi = np.sum((y - y_pred_multi)**2)
print(f"RSS: {RSS_multi:.2f}")
```

### From the RSS claculated in single and multible linear regressions, can you explain why using 'mother'and'father' height features did not reduce RSS? 

Even though we added `father` and `mother` separately, the **midparentHeight** feature is already a 
linear combination of father and mother heights (often defined as 

$$
\text{midparentHeight} \;=\; \frac{\text{father} + \text{mother}}{2}
$$

sometimes with a small adjustment).

This means `father` and `mother` dont provide much new independent information beyond what is already 
captured by `midparentHeight`. Because of this strong collinearity, adding them does not reduce RSS  
the simple regression with `midparentHeight` already explains almost all of the variation in `childHeight`.
