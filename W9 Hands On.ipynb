## Exercise 1: Logistic Regression [12 Marks]

### Import the required libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

### Run the following cell to prepare your data.

```python
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)

X = iris.data[["petal length (cm)"]].values

y = iris.target_names[iris.target] == 'setosa'
```

```python
X
```

### Draw a histplot for the input variable X [1 Mark]

```python
sns.histplot(X)
plt.show()
```

```python
y
```

### Draw a countplot for the dependent variable y [1 Mark]

```python
sns.countplot(x=y, hue=y, legend=False)
plt.show()
```

### Train/Test split data with test_size=0.25, random_state=42 [2 Marks]

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

### Create a logistic regression model [2 Marks]

```python
log_reg = LogisticRegression()
print(log_reg)
```

### Fit the logistic regression model to the training data [2 Marks]

```python
log_reg.fit(X_train, y_train)
```

### Predict the output for the training and test data [2 Marks]

```python
y_train_pred = log_reg.predict(X_train)
y_test_pred = log_reg.predict(X_test)
```

### Evaluate the accuracy of prediction in training and test data [2 Marks]

```python
acc_train = accuracy_score(y_train, y_train_pred)
print("Training prediction accuracy:", acc_train)
```

```python
acc_test = accuracy_score(y_test, y_test_pred)
print("Testing prediction accuracy:", acc_test)
```

## Exercise 2: Logistic Regression (Multi-Variables) [8 Marks]

### Run the following cell to prepare your data.

```python
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)

X = iris.data[['sepal length (cm)','sepal width (cm)']].values

y = iris.target_names[iris.target] == 'virginica'
```

```python
X
```

### Draw a histplot for each of the input variables x1, x2 [1 Mark]

```python
sns.histplot(X[:,0])
```

```python
sns.histplot(X[:,1])
```

### Draw a scatter plot x1 versus x2 [1 Mark]

```python
plt.scatter(X[:,0],X[:,1])
```

```python
y
```

### Draw a countplot for the dependent variable y [1 Mark]

```python
sns.countplot(x=y, hue=y, legend=False)
```

### Draw a scatter plot x1 versus x2 colored by the class label

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[y==1, 0], X[y==1, 1], color='red', label='Virginica')
plt.scatter(X[y==0, 0], X[y==0, 1], color='blue', label='Not Virginica')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal length (cm)')
plt.title('Iris Dataset: sepal length vs. sepal length')
plt.legend()
plt.show()
```

### Train/Test split data with test_size=0.25, random_state=42 [1 Mark]

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

### Create a logistic regression model [1 Marks]

```python
log_reg = LogisticRegression()
print(log_reg)
```

### Fit the logistic regression model to the training data [1 Mark]

```python
log_reg.fit(X_train, y_train)
```

### Predict the output for the training and test data [1 Mark]

```python
y_train_pred = log_reg.predict(X_train)
```

```python
y_test_pred = log_reg.predict(X_test)
```

### Evaluate the accuracy of prediction on training and test data [1 Mark]

```python
training_accuracy = accuracy_score(y_train, y_train_pred)
print("Training prediction accuracy:", training_accuracy)
```

```python
testing_accuracy = accuracy_score(y_test, y_test_pred)
print("Testing prediction accuracy:", testing_accuracy)
```

```python
coef = log_reg.coef_[0]
intercept = log_reg.intercept_[0]

# Generate a range of values for the x-axis (petal width)
x_values = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)

# Calculate the corresponding y-values (petal length) for the decision boundary
y_values = -(coef[0] * x_values + intercept) / coef[1]

plt.figure(figsize=(8, 6))
plt.scatter(X[y==1, 0], X[y==1, 1], color='red', label='Virginica')
plt.scatter(X[y==0, 0], X[y==0, 1], color='blue', label='Not Virginica')
plt.plot(x_values, y_values, color='green', label='Decision Boundary')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal length (cm)')
plt.title('Iris Dataset: sepal length vs. sepal length')
plt.legend()
plt.ylim([1.75,4.5])
plt.show()
```
